# Preparing the data for the model. Balancing the data for the model, taking a sample of the legit and fraud transactions. Then splitting the data info Features and Targets. Finally splitting the data into Training and Testing.

```python
legit_sample = legit.sample(n = 200)
fraud_sample = fraud.sample(n = 200)
sample_df = pd.concat([legit_sample, fraud_sample], axis = 0)
sample_df.groupby('Class').mean()
```

```python
X = sample_df.drop(columns = 'Class', axis = 1)
Y = sample_df['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
print(X.shape, X_train.shape, X_test.shape)
```

# Training the model and evaluation

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(training_data_accuracy)

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(test_data_accuracy)
```
